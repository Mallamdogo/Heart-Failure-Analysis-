```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(Records.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(Records['age'], bins=20, kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Ejection Fraction
plt.figure(figsize=(8, 6))
sns.histplot(Records['ejection_fraction'], bins=15, kde=True, color='green')
plt.title('Distribution of Ejection Fraction')
plt.xlabel('Ejection Fraction')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Serum Creatinine
plt.figure(figsize=(8, 6))
sns.histplot(Records['serum_creatinine'], bins=15, kde=True, color='orange')
plt.title('Distribution of Serum Creatinine')
plt.xlabel('Serum Creatinine')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot for Death Events
plt.figure(figsize=(8, 6))
sns.countplot(x='DEATH_EVENT', data=Records, palette='pastel')
plt.title('Distribution of Death Events')
plt.xlabel('Death Event')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Survived', 'Deceased'])
plt.show()
```

```python
# Box plot for Age vs. Death Event
plt.figure(figsize=(8, 6))
sns.boxplot(x='DEATH_EVENT', y='age', data=Records, palette='pastel')
plt.title('Age Distribution by Death Event')
plt.xlabel('Death Event')
plt.ylabel('Age')
plt.xticks(ticks=[0, 1], labels=['Survived', 'Deceased'])
plt.show()
```

```python
# Scatter plot for Age vs. Serum Creatinine
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='serum_creatinine', data=Records, hue='DEATH_EVENT', palette='pastel')
plt.title('Age vs. Serum Creatinine')
plt.xlabel('Age')
plt.ylabel('Serum Creatinine')
plt.legend(title='Death Event', labels=['Survived', 'Deceased'])
plt.show()
```

```python

# Pairplot for selected features
selected_features = ['age', 'ejection_fraction', 'serum_creatinine', 'serum_sodium', 'time', 'DEATH_EVENT']
sns.pairplot(Records[selected_features], hue='DEATH_EVENT', palette='pastel')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
# Violin plot for Ejection Fraction by Death Event
plt.figure(figsize=(8, 6))
sns.violinplot(x='DEATH_EVENT', y='ejection_fraction', data=Records, palette='pastel')
plt.title('Ejection Fraction by Death Event')
plt.xlabel('Death Event')
plt.ylabel('Ejection Fraction')
plt.xticks(ticks=[0, 1], labels=['Survived', 'Deceased'])
plt.show()
```
